```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Assuming 'df' is your DataFrame containing the 'Purchased items' column
# Convert the 'Purchased items' column to a list of lists
transactions = df['Purchased items'].str.split(',').tolist()

# Convert the transaction data to the required format for the Apriori algorithm
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_transformed = pd.DataFrame(te_ary, columns=te.columns_)

# Apply the Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_transformed, min_support=0.01, use_colnames=True)

# Generate all association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display all the association rules
print(rules)

```

```python
# Setting the thresholds for support, confidence, and lift
min_support = 0.01
min_confidence = 0.5
min_lift = 1.2

# Filter the rules based on the thresholds
filtered_rules = rules[(rules['support'] >= min_support) &
                        (rules['confidence'] >= min_confidence) &
                        (rules['lift'] >= min_lift)]

# Display the filtered rules
print(filtered_rules)
```

```python
#These are the association rules that meet the specified thresholds:

#(1) If a customer buys eggs and ground beef, they are 50.67% likely to also buy mineral water.
#This rule has a lift of 2.13, indicating a significant relationship between the antecedent and consequent.

#(2)If a customer buys milk and ground beef, they are 50.30% likely to also buy mineral water.
# This rule has a lift of 2.11, indicating a significant relationship between the antecedent and consequent.

#These rules have a support of at least 1%, confidence of at least 50%, and lift of at least 1.2.






```
